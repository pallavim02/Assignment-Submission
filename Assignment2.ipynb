```python
import plotly.express as px
```

```python
df=px.data.iris()
```

```python
df.head()
```

```python
fig=px.scatter(df,x="sepal_length",y="sepal_width",color="species")
fig.show()
```

```python
fig=px.scatter(df,x="sepal_length",y="sepal_width",color="species",marginal_y="violin",marginal_x='box',trendline='ols',
               template="simple_white")
fig.show()
```

```python
df=px.data.tips()
df.head(10)
```

```python
fig=px.bar(df,x='sex',y='total_bill',color='smoker',barmode="group")
fig.show()
```

```python
df=px.data.medals_long()

fig=px.bar(df,x="medal",y='count',color='nation',pattern_shape='nation',pattern_shape_sequence=[".","x","+"])
fig.show()
```

```python
df=px.data.tips()
fig=px.violin(df,y="tip",x="smoker",color="sex",box=True,points="all",hover_data=df.columns)
```

```python
fig.show()
```

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
fig=px.line_3d(flights,x='year',y='month',z='passengers',color='year')
fig
```

```python
flights=sns.load_dataset("flights")
fig=px.scatter_3d(flights,x='year',y='month',z='passengers',color='year')
fig.show()
```

```python
df=px.data.iris()
fig=px.scatter_matrix(df,dimensions=["sepal_width","sepal_length","petal_width","petal_length"],color='species')
fig.show()
```

```python
df1=px.data.gapminder().query("year==2007")
fig=px.sunburst(df1,path=['continent','country'],values='pop',color='lifeExp',hover_data=['iso_alpha'])
fig.show()
```

```python
df1.head()
```

```python
df=px.data.iris()
fig=px.parallel_coordinates(df,color='species_id',labels={"species_id": "Species","Sepal_width":"sepal width",
                                                          "sepal_length":"sepal length",},
                           color_continuous_midpoint=2)
fig.show()
```
